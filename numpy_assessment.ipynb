```python

df=pd.read_csv("E:/Data Analysis datasets/historical_automobile_sales.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt

df_line = df.groupby('Year')['Automobile_Sales'].mean()

plt.figure(figsize=(10, 5),facecolor='aqua')
df_line.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Automobile Sales Fluctuation from Year to Year')
plt.xticks(sorted(df['Year'].unique()), rotation=45)

plt.tight_layout()
```

```python
recession_data = df[df['Recession'] == 1]

df_recession = recession_data.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()
df_pivot = df_recession.pivot(index='Year', columns='Vehicle_Type', values='Automobile_Sales')

plt.figure(figsize=(12, 7))
df_pivot.plot(kind='line', marker='o', ax=plt.gca())

plt.title('Vehicle Sales Trends by Type During Recession Periods')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.legend(title='Vehicle Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
```

```python
import seaborn as sns
df_grouped = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=df_grouped, x='Vehicle_Type', y='Automobile_Sales', hue='Recession',palette='inferno')

plt.title('Average Automobile Sales by Vehicle Type: Recession vs. Non-Recession')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Automobile Sales')
plt.legend(title='Recession', labels=['No', 'Yes'])
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7), sharey=False)

sns.barplot(x='Vehicle_Type', y='Automobile_Sales', data=df[df['Recession'] == 0], ax=ax1, estimator='mean', palette='inferno')
ax1.set_title('Average Sales by Vehicle Type: Non-Recession Period')
ax1.set_xlabel('Vehicle Type')
ax1.set_ylabel('Average Automobile Sales')
ax1.tick_params(axis='x', rotation=45)

sns.barplot(x='Vehicle_Type', y='Automobile_Sales', data=df[df['Recession'] == 1], ax=ax2, estimator='mean',palette="magma")
ax2.set_title('Average Sales by Vehicle Type: Recession Period')
ax2.set_xlabel('Vehicle Type')
ax2.set_ylabel('Average Automobile Sales')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
```
